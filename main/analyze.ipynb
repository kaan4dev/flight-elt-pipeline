```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/kaancakir/Library/CloudStorage/OneDrive-KisÌ§isel/data/projects/flight-elt-pipeline/flights_transformed.csv')
```

```python
average_delay_by_airline = df.groupby('airline_name')['departure_delay'].mean().sort_values(ascending=False)
print('Average delay time of airlines (per minute): ')
print(average_delay_by_airline)
```

```python
flight_status_counts = df['flight_status'].value_counts()
print("Distrubution Of Flight Status: ")
print(flight_status_counts)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(average_delay_by_airline.index, average_delay_by_airline.values, color='skyblue')
plt.xticks(rotation=45)
plt.title('Average delay time of airlines')
plt.xlabel('Airline')
plt.ylabel('Average Delay per minute')
plt.tight_layout()
plt.show()
```
